```python
#separating data for analysis
normal=credit_card[credit_card.Class==0]
fraud=credit_card[credit_card.Class==1]
print(normal.shape)
print(fraud.shape)
```

dealing with unbalanced data using under-sampling

build a dataset containing similar distribution of normal transaction and fraudulent transaction

Number of fraudulent data =102

```python
normal_sample=normal.sample(n=102)
```

now we concatinate two dataframes 1 is normal_sample and 2nd is fraud transaction

```python
new=pd.concat([normal_sample,fraud],axis=0)
new.head()
```

```python
new.tail()
```

```python
new['Class'].value_counts()
```

```python
new.groupby('Class').mean()
```

spliting data into features and targets

```python
x=new.drop(columns='Class',axis=1)
y=new['Class']
```

```python
print(x)
```

```python
print(y)
```

split data into training and testing data

test size represent amount of data if we are writing 0.2 it means 80% is train data and 20% is test data

stratify=Y is start data based on Y
it distributes Y evenly in training and testing data

random state is for randomness

---



```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
print(x.shape,x_train.shape,x_test.shape)
```

model training

we are using logistic regression method 
it is used for binary classification  

```python
model=LogisticRegression()
```

```python
#training data
model.fit(x_train,y_train)
```

model evalution

find performance of our model by ACCURACY SCORE

```python
#accuracy on training data
x_train_prediction = model.predict(x_train)
training_acc=accuracy_score(x_train_prediction,y_train)
print(" accuracy on training data",training_acc)
```

```python
#accuracy on test data
x_test_prediction = model.predict(x_test)
testing_acc=accuracy_score(x_test_prediction,y_test)
print(" accuracy on testing data",testing_acc)
```

if our training accuracy is higher and testing accuracy is lower then our model is overfitted

if our training accuracy is lower and testing accuracy is higher then our model is underfitted

but in our case both accuracy are similar so we can say that our model is perfectly trained.
